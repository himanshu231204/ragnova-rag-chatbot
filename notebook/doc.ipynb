# Data Ingestion

```python
from langchain_community.document_loaders import TextLoader, PyPDFLoader ,DirectoryLoader
```

```python
loader=PyPDFLoader("book.pdf")
```

```python
#load the documents
doc=loader.load()
```

```python
#content of the  document & metadata of the  document
print(doc[100].page_content)
print(doc[188].metadata) #output: page_content='This is the content of the first page
```

```python
print(len(doc)) # page in the pdf
```

```python
# create a directory loader object
lod = DirectoryLoader(
    path="data",
    glob="**/*.pdf",
    loader_cls=PyPDFLoader
)
```

```python
doc1 = list(lod.lazy_load())

print(f"Loaded documents: {len(doc1)}")
for d in doc1[:2]:
    print(d.page_content[:300])
    print(d.metadata)
```

```python
if doc1:
    print(doc1[0].page_content)
else:
    print("No documents loaded. Check path/glob.")
```
